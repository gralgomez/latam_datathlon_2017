```python
import numpy as np
import os
import matplotlib
import pandas as pd
```

```python
#read in whatever datafile you want
def read_datafunc(datafile):
    read_data = pd.read_csv(datafile)
    return read_data
```

```python
#summarize the size and columns, of the data
def summarize_data(datafile):
    read_data = read_datafunc(datafile)
    print(datafile)
    print('Dimensions:', read_data.shape)
    print('Columns:', list(read_data.columns))
    return read_data   

df = summarize_data('../R/ind_salud_correctID.csv')
```

```python
#removing columns that have highly sparse data
sparsity_threshold = .5
df_nonsparse_columns = df.loc[:,(df.isnull().sum()/len(df)) < sparsity_threshold]

#look at the new shape of the dataset
print(df_nonsparse_columns.shape)
print(df_nonsparse_columns.columns)
```

```python
keep = df_nonsparse_columns.dropna(subset=['yea2005', 'yea2006', 'yea2007', 'yea2008', 'yea2009', 'yea2010'], thresh=3)
keep = keep[['idindicador', 'nomindicador', 'idunidad', 'nomunidad',
       'iddepto', 'nomdepto', 'nommpio', 'yea2005', 'yea2006', 'yea2007',
       'yea2008', 'yea2009', 'yea2010']]
keep.rename(columns = {'yea2005':'2005'}, inplace = True)
ttdiesti = keep.loc[(keep['nomdepto'] != 'extranjeros') & (keep['idindicador'] == 'ttdiesti'), :]

ttdiesti_melted = pd.melt(ttdiesti, id_vars=['idindicador', 'nomindicador', 'idunidad', 'nomunidad',
       'iddepto', 'nomdepto', 'nommpio'])

ttdiesti_final = ttdiesti_melted[['iddepto','value']]

print(ttdiesti_final)

pd.DataFrame.to_csv('..data/ttdiesti.csv')
```
