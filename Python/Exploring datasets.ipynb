```python
keep = df_nonsparse_columns.dropna(subset=['yea2005', 'yea2006', 'yea2007', 'yea2008', 'yea2009', 'yea2010'], thresh=3)
keep = keep[['idindicador', 'nomindicador', 'idunidad', 'nomunidad',
       'iddepto', 'nomdepto', 'nommpio', 'yea2005', 'yea2006', 'yea2007',
       'yea2008', 'yea2009', 'yea2010']]

ttdiesti = keep.loc[(keep['nomdepto'] != 'extranjeros') & (keep['idindicador'] == 'ttdiesti'), :]

ttdiesti_melted = pd.melt(ttdiesti, id_vars=['idindicador', 'nomindicador', 'idunidad', 'nomunidad',
       'iddepto', 'nomdepto', 'nommpio'], var_name='year')

ttdiesti_2010 = ttdiesti_melted.loc[ttdiesti_melted['year'] == 'yea2010', :]
print(ttdiesti_2010)
ttdiesti_final = ttdiesti_2010[['iddepto','value']]

print(ttdiesti_final)

ttdiesti_final.to_csv('ttdiesti.csv')
```

```python
keep_2010 = keep[['idindicador', 'iddepto', 'yea2010']].dropna().drop_duplicates()
grouped = keep_2010.groupby(['iddepto', 'idindicador']).agg('median').reset_index().drop_duplicates()
print(grouped['idindicador'].drop_duplicates())

import numpy as np
bydepto2010 = pd.pivot_table(grouped, index='iddepto', values='yea2010', columns='idindicador')
bydepto2010.corr().to_csv('2010_indicador_corr_by_depto.csv')
```
